```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['Basic', 'Standard', 'Premium']
values = [40, 31, 29]

# Custom color palette for the pie chart slices
colors = ['cyan', 'pink', 'pink']

# Create a figure and axis
fig, ax = plt.subplots()

# Define properties for the pie chart wedges, including edgecolor
wedge_props = {'linewidth': 3, 'edgecolor': 'grey'}

# Plot the data as a pie chart with custom colors and wedge properties
ax.pie(values, labels=categories, autopct='%1.1f%%', startangle=90, colors=colors, wedgeprops=wedge_props)

# Add a title
ax.set_title('Netflix Userbases')

# Display the chart
plt.show()

# Devices Data

fig, ax = plt.subplots()

fruits = ['Laptop', 'Other', 'Tablet']
counts = [25, 49, 25]
bar_labels = ['Laptop', 'Other', 'Tablet']
bar_colors = ['tab:blue', 'tab:red', 'tab:blue']
bar_edge_colors = ['black', 'black', 'black']  # Specify edge colors for each bar
bar_edge_width = 2  # Specify the width of the bar edges

# Plot the data as a bar plot with color and size for edges
bars = ax.bar(fruits, counts, label=bar_labels, color=bar_colors, edgecolor=bar_edge_colors, linewidth=bar_edge_width)

# Add labels and title
ax.set_ylabel('Use Data')
ax.set_title('Netflix Devices')
ax.legend(title='Netflix')

# Add custom labels for each bar
for bar, label in zip(bars, bar_labels):
    height = bar.get_height()
    ax.annotate(f'{height}', xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')

plt.show()


# Count of countries

import matplotlib.pyplot as plt

# Data
categories = ['United States', 'Canada', 'Mexico', 'Brasil', 'Spain', 'United Kingdom',
              'France', 'Germany', 'Italy', 'Australia']
values = [451, 317, 183, 183, 451, 183, 183, 183, 183, 183]

# Create a figure and axis with larger figure size
fig, ax = plt.subplots(figsize=(10, 6))

# Set custom colors for the bars
colors = ['blue', 'green', 'lime', 'lime', 'blue', 'lime', 'lime', 'lime', 'lime', 'lime']
# Set custom width for the edge lines
edge_width = 1.7

# Plot the data as a bar chart with custom colors and edge lines
ax.bar(categories, values, edgecolor='black', color=colors, linewidth=edge_width)

# Add labels and title
ax.set_xlabel('Countries')
ax.set_ylabel('Userbases')
ax.set_title('Netflix')

# Rotate the x-axis labels to make them more separate
plt.xticks(rotation=45, ha='right')

# Display the chart
plt.tight_layout()
plt.show()

# Gender

# Generate some example data
np.random.seed(42)
data1 = np.random.normal(loc=0, scale=1, size=100)
data2 = np.random.normal(loc=2, scale=1, size=100)

# Data for the violin plot
data = [data1, data2]
labels = ['Male', 'Female']

# Colors for the violins
colors = ['blue', 'pink']

# Edge colors for the violins
edge_colors = ['black', 'black']

# Create a figure and axis
fig, ax = plt.subplots()

# Set the positions of the violins side by side
positions = [1, 1]

# Plot the data as a violin plot with custom colors and edge colors
violins = ax.violinplot(data, positions=positions, showmedians=True)

# Set custom colors and edge colors for the violins
for i, (violin, color, edge_color) in enumerate(zip(violins['bodies'], colors, edge_colors)):
    violin.set_facecolor(color)
    violin.set_edgecolor(edge_color)

# Add labels and title
ax.set_xlabel('50% / 50%')
ax.set_ylabel('Values')
ax.set_title('Gender')

# Set the x-axis ticks and labels
ax.set_xticks(positions)
ax.set_xticklabels(labels)

# Display the chart
plt.show()
```
